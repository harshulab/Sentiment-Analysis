# SENTIMENT ANALYSIS ON RAW COMMENTS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('GBcomments.csv', error_bad_lines=False)
```

```python
df
```

```python
# !pip install textblob
```

```python
from textblob import TextBlob
```

```python
df.dropna(inplace=True)
```

```python
# print(TextBlob('Best').sentiment.polarity)
# print(TextBlob('Neutral').sentiment.polarity)
# print(TextBlob('Worst').sentiment.polarity)
```

```python
data = df[:2000]
data
```

```python
for i in range(len(data)):
    data['polarity'].iloc[i] = TextBlob(df['comment_text'].iloc[i]).sentiment.polarity
data
```

```python
polarity_one = data[data['polarity']==1]
polarity_one
```

# TO ANALYSE TRENDING TAGS ON YOUTUBE

```python
dftag = pd.read_csv('USvideos.csv', error_bad_lines=False)
dftag
```

```python
# dftag['split_tag']=0
# dftag
```

```python
for i in range(len(dftag)):
    dftag['split_tag'].iloc[i] = dftag['tags'].iloc[i].split('|')
dftag
```

```python
dftag['split_tag'].iloc[0]

freq={}
for i in range(len(dftag)):
    for tag in dftag['split_tag'].iloc[i]:
        if tag in freq:
            freq[tag]+=1
        else:
            freq[tag]=1
max_key = max(freq, key=freq.get)
print(max_key)
print(freq)
```

### To generate wordcloud of trending tags

```python
tags_for_cloud = ' '.join(dftag['tags'])
tags_for_cloud
import re #for easy removal of character not needed
tags_for_cloud = re.sub('[^a-zA-Z]', ' ', tags_for_cloud) #everything except a-z and A-Z should be sub(stituted) with ' '
tags_for_cloud = re.sub(' +', ' ', tags_for_cloud)
tags_for_cloud
```

```python
cloud2 = WordCloud(width=1000, height=500, stopwords=set(STOPWORDS)).generate(tags_for_cloud)
plt.figure(figsize=(15,5))
plt.imshow(cloud2)
```
