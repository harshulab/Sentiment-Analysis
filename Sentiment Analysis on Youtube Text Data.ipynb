# TO ANALYSE TRENDING TAGS ON YOUTUBE

```python
dftag = pd.read_csv('USvideos.csv', error_bad_lines=False)
dftag
```

```python
dftag['split_tag']=0
dftag
```

```python
for i in range(len(dftag)):
    dftag['split_tag'].iloc[i] = dftag['tags'].iloc[i].split('|')
dftag
```

```python
dftag['split_tag'].iloc[0]

freq={}
for i in range(len(dftag)):
    for tag in dftag['split_tag'].iloc[i]:
        if tag in freq:
            freq[tag]+=1
        else:
            freq[tag]=1
max_key = max(freq, key=freq.get)
print(max_key)
print(freq)
```

### To generate wordcloud of trending tags

```python
tags_for_cloud = ' '.join(dftag['tags'])
tags_for_cloud
import re #for easy removal of character not needed
tags_for_cloud = re.sub('[^a-zA-Z]', ' ', tags_for_cloud) #everything except a-z and A-Z should be sub(stituted) with ' '
tags_for_cloud = re.sub(' +', ' ', tags_for_cloud)
tags_for_cloud
```

```python
cloud2 = WordCloud(width=1000, height=500, stopwords=set(STOPWORDS)).generate(tags_for_cloud)
plt.figure(figsize=(15,5))
plt.imshow(cloud2)
```

```python
import emoji 
```

```python
emoji_str = []
for comm in df['comment_text']:
    for c in comm:
        if c in emoji.UNICODE_EMOJI['en']:
            emoji_str.append(c)
        
emoji_str
```

```python
emoji_freq={}
for i in set(emoji_str):
    emoji_freq[i] = emoji_str.count(i)

emoji_freq = {k: v for k, v in sorted(emoji_freq.items(), key=lambda item: item[1], reverse=True)}
emoji_freq
```

```python
#we need to enumerate this dictionary... but dictionaries in older version so fpython are not enumerated
#newer version support enumerated dict, more on that later
#so we convert this dict into df and then plot it

emoji_df = pd.DataFrame(emoji_freq.items(), columns=['emoji_symbol', 'emoji_freq'])
emoji_df_small = emoji_df.head(10)
emoji_df_small
```

```python
import matplotlib.pyplot as plt

plt.bar(range(10), emoji_df_small['emoji_freq'])
plt.xticks(range(10), emoji_df_small['emoji_symbol'])
plt.show()
```
